### Activity 3: Commas to Dot replacement

The  values in the `CO(GT), C6H6(GT), T,	RH` and	`AH` columns contain the commas in them. Let's replace the commas with periods (or dots).

```
#replace the commas with periods in a Pandas series.
def replace_comma(b):
    list1 = []
    for i in b:
        list1.append(float(str(i).replace(",",".")))
    return pd.Series(list1)
```

```
replace_comma(["hey,,,,","who,,,"])
```

```
replace_comma([1,5])
```

```
replace_comma(air_df['CO(GT)'])
```

```
#Apply the 'replace_comma()' function on the ''CO(GT)', 'C6H6(GT)', 'T', 'RH' and 'AH' columns.
list2 = ['CO(GT)', 'C6H6(GT)', 'T', 'RH', 'AH' ]
for i in list2:
    air_df[i] = replace_comma(air_df[i])
```

### Activity 3: The `loc[]` Function

Now, let's replace the garbage values with the median of  values. For this, we need to use the `loc[]` function from Pandas library. It is an extension to the `iloc[]` function. The `iloc[]` function takes only the row and column indices as inputs. But in addition to the row and column indices, the `loc[]` function also accepts the array of row indices and column indices as inputs. It also accepts the specific column names as inputs.



```
#Count the number of rows containing '2004' and '2005' year values.
air_df['Year'].value_counts()
```

```
# Create a new DataFrame containing records for the year 2004.
air_df_2004 = air_df.loc[air_df['Year'] == 2004, air_df.columns]
air_df_2004.head()
```

```
# Calculate the median values for the columns having indices between 1 and -4 (excluding -4) for the year 2004.
air_df.loc[air_df["Year"] == 2004, air_df.columns[1:-4]].median()
```

```
air_df.iloc[[2,3,4],[1,2]]
```

```
# Calculate the median values for the columns  for the year 2005.
air_df.loc[air_df['Year'] == 2005, air_df.columns[1:-4]].median()
```

```
# Replace the -200 value with the median values for the 2004 year DataFrame.
for i in air_df_2004.columns[1:-4]:
 median = air_df_2004[i].median()
 air_df_2004[i] = air_df_2004[i].replace(to_replace= -200, value= median)
```

```
# Create a new DataFrame containing records for the year 2004.
air_df_2005= air_df.loc[air_df['Year'] == 2005, air_df.columns]
air_df_2005.head()
```

```
# Repeat the same for the 2005 year DataFrame.
for i in air_df_2005.columns[1:-4]:
 median = air_df_2005[i].median()
 air_df_2005[i] = air_df_2005[i].replace(to_replace= -200, value= median)
```
