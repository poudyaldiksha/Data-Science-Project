## About Dataset

**Finding Exoplanets Principle**



NASA placed the Kepler telescope in space to measure the brightness of stars in distant galaxies. When a planet passes between the telescope and a star, the recorded brightness decreases. When the planet moves behind the star, the brightness increases. This method of detecting exoplanets through the brightness of the light emitted by a star is called the Transit Method.

When plotting brightness (vertical axis) against time (horizontal axis), the brightness of the star recorded by the telescope shows a wave-like pattern, indicating the presence of at least one planet.


### Activity 4: Slicing a DataFrame Using the `iloc[]` Function

Extract specific rows and columns from the DataFrame using the` iloc[]` function.

The **iloc** indexer in Pandas is a powerful tool used to select rows and columns from a DataFrame by their integer position (i.e., their index position). It stands for "integer location" and allows for both row and column indexing.

**Syntax:**

`dataframe_name.iloc[row_position_start : row_position_end, column_position_start : column_position_end]`

In this syntax:

- `row_position_start` denotes the position of the row in the DataFrame **starting** from whose values you want to take in the new Pandas series or DataFrame.
- `row_position_end` denotes the position of the row in the DataFrame till whose values you want to take in the new Pandas series or DataFrame.
- `column_position_start` denotes the position of the column in the DataFrame **starting** from whose values you want to take in the new Pandas series or DataFrame.
- `column_position_end` denotes the position of the column in the DataFrame till whose values you want to take in the new Pandas series or DataFrame.

### Activity 7: Column Operations In pandas dataframe

#### Renaming columns
**Rename a Single Column:** To rename a single column, use the rename method.

````python
df = pd.DataFrame({
    'city': ['New York', 'Los Angeles', 'Chicago'],
    'population': [8175133, 3792621, 2695598]
})

````

````python
df.rename(columns={'city': 'city_name'})
print(df)
````

**Rename Multiple Columns:** To rename multiple columns, specify each column in the dictionary.

````python

df = pd.DataFrame({
    'city': ['New York', 'Los Angeles', 'Chicago'],
    'population': [8175133, 3792621, 2695598]
})

````

````python
df.rename(columns={'city': 'city_name', 'population': 'population_count'}, inplace=True)
print(df)
````

### Activity 8: Apply function to column names
**Make All Column Names Uppercase:**
You can convert all column names to uppercase using list comprehension.

````python
df = pd.DataFrame({
    'city': ['New York', 'Los Angeles', 'Chicago'],
    'population': [8175133, 3792621, 2695598]
})

df.columns = [col.upper() for col in df.columns]
print(df)
````

### Activity 9: Applying Functions to Column Data


**Convert Column Values to Uppercase:**
Use the map function to apply a lambda function to a column.


````python
df = pd.DataFrame({
    'city': ['New York', 'Los Angeles', 'Chicago'],
    'population': [8175133, 3792621, 2695598]
})

df['city'] = df['city'].map(lambda city: city.upper())
print(df)
````

**Key Points:**

- **Map** is a built-in function in Python that allows you to iterate over an iterable (like a list or column in a DataFrame) and apply a function to each element.
- **Lambda** functions provide a way to define small, anonymous functions in Python without the need for a separate def statement. They are especially useful when you need a short, one-line function for a specific task within another function like map.

### Activity 10: Creating Derived Columns
**Add a New Column Based on Existing Data:**
You can create new columns using existing data in the DataFrame.And it's very common to add new columns using derived data. You just need to assign to a new column:

````python
df = pd.DataFrame({
    'city': ['New York', 'Los Angeles', 'Chicago'],
    'population': [8175133, 3792621, 2695598]
})

df['population_in_millions'] = df['population'] / 1000000
print(df)
````

###Activity 11: Reordering Columns

**Change Column Order:**
Reorder columns by reassigning the DataFrame with columns in the desired order.

````python
df = pd.DataFrame({
    'city': ['New York', 'Los Angeles', 'Chicago'],
    'population': [8175133, 3792621, 2695598]
})

df = df[['population', 'city']]
print(df)
````

###Activity 12: Dropping Columns

**Drop a Single Column:**
Use `df.drop(columns=['column_name'])` to drop a single column.

````python

df = pd.DataFrame({
    'city': ['New York', 'Los Angeles', 'Chicago'],
    'population': [8175133, 3792621, 2695598]
})

df = df.drop(columns=['population'])
print(df)
````

**Drop Multiple Columns:**
Pass multiple column names to the columns argument to drop multiple columns.



````python
df = pd.DataFrame({
    'city': ['New York', 'Los Angeles', 'Chicago'],
    'population': [8175133, 3792621, 2695598]
})

df = df.drop(columns=['city', 'population'])
print(df)
````

### Activity 12: Adding New Columns

**Append a New Column:**
Add a new column by creating a Series and assigning it to the DataFrame.

````python
df = pd.DataFrame({
    'city': ['New York', 'Los Angeles', 'Chicago'],
    'population': [8175133, 3792621, 2695598]
})

states = pd.Series(['NY', 'CA', 'IL'])
df['state'] = states
print(df)
````

### Activity 13: Inserting Columns at Specific Positions

**Insert a Column at a Specific Position:**
Use `df.insert(position, column_name, data)` to insert a column at a specific position.

````python

df = pd.DataFrame({
    'city': ['New York', 'Los Angeles', 'Chicago'],
    'population': [8175133, 3792621, 2695598]
})

gender = pd.Series(['female', 'male', 'male'])
position = 1
column_name = 'gender'

df.insert(position, column_name, gender)
print(df)
````
