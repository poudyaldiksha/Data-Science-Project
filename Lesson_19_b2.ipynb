### Activity 3: Applying Functions to Column Data and `map()`


**Convert Column Values to Uppercase:**
Use the map function to apply a lambda function to a column.


```python
df = pd.DataFrame({
    'city': ['New York', 'Los Angeles', 'Chicago'],
    'population': [8175133, 3792621, 2695598]
})
```

```python
#Activity
df["city"] = df["city"].map(lambda city : city.upper())
df
```

**Key Points:**

- **Map** is a built-in function in Python that allows you to iterate over an iterable (like a list or column in a DataFrame) and apply a function to each element.
- **Lambda** functions provide a way to define small, anonymous functions in Python without the need for a separate def statement. They are especially useful when you need a short, one-line function for a specific task within another function like map.

### Activity 10: Converting Columns to Dates/Datetimes
**Convert Column to Date/Datetime**
Use `pd.to_datetime()` to convert a column to a `datetime` values  so that later we can sort the DataFrame in the chronological order. A `datetime` value is a combination of date and time.

**Syntax:** `pd.to_datetime(series)`

where the `series` is the Pandas series whose values are to be converted into `datetime` values.

The default format of a `datetime` value is `YYYY-MM-DD` where `YYYY` denotes year, `MM` denotes month and `DD` denotes day.

- `HH` denotes hours

- `MM` denotes minutes

- `SS` denotes seconds

```python
df = pd.DataFrame({
    'city': ['New York', 'Los Angeles', 'Chicago'],
    'established': ['1898-07-04', '1781-09-04', '1837-03-04']
})

df['established'] = pd.to_datetime(df['established'])
print(df)
```

#Map function


The strip function  isn't working because map doesn't directly apply functions that modify the original object (strings in this case) within a pandas DataFrame.

Here's what's happening:

- **map with strip:** You're trying to use map with the strip function on the names column. However, map is designed to apply a function element-wise to a Series or column, often for basic transformations like replacements. It doesn't modify the original objects in place.

- **String Immutability:** Strings in Python are immutable, meaning you cannot directly change a string's content after creation. Any modifications create a new string object.

- **Solution:** Apply strip with apply

To achieve the desired result of removing leading/trailing spaces from names, you can use the apply function with the strip function:



```python
import pandas as pd

df = pd.DataFrame({
    'names': ['Alice', ' Bob', ' Charlie ']
})

df['name_length'] = df['names'].apply(lambda x: x.strip())  # Apply strip with lambda function
print(df)
```
