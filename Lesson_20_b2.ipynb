### Activity 4: Parsing `datetime` Values - `date(), strftime()` & `time()` Functions

Let's learn how to extract date, day, month, year, time, hours, minutes and seconds from a `datetime` value.

**The `date()` Function**

In general, a `datetime` value is a `Timestamp` object. It holds both date and time.

```
# Get the first 'datetime' value
timestamp_0 = date_time[0]
timestamp_0
```

To get a date (i.e., `datetime.date` object) from a timestamp, apply the `date()` function.

```
# Get date from the timestamp stored in the 'timestamp_0' variable
# Print the value and its data-type.
date_0=timestamp_0.date()
date_0
```

**The `day, month` & `year` Attributes**

To get day, month and year from a `datetime.date` value apply the `day, month` and `year` attributes.

The output will be day, month and year as integer values.

```
# Get the day, month and year attributes from the date stored in the 'date_0' variable
print(date_0.year)
print(date_0.month)
print(date_0.day)
```

**The `strftime()` Function**

To convert a `datetime.date` value into a string value, use the `strftime()` function. You can also format the date in different formats using the `strftime()` function.

Let's format the date stored in the `date_0` variable in the `10-Mar-2004` format. To do this, you will have to pass the following input to the `strftime()` function.

`'%d-%b-%Y'`

It is a combination of three format codes. They are:

1. `%d` format code denotes day of the month as a zero-padded decimal number. E.g., 01, 02, 03 etc.

2. `%b` format code denotes abbreviated month name of month in the `datetime.date` value. E.g., Jan Feb, March etc.

3. `%Y` format code denotes year with century as a decimal number. E.g, 2019, 2020, 2021 etc.

The hyphen `-` used here just to separate the day, month and year values. You can separate them with different characters.

To get a list of all the `strftime()` format codes, click on the link provided below and then go to the **`strftime()` and `strptime() Format Codes`** section.

https://docs.python.org/3/library/datetime.html#strftime-and-strptime-format-codes



```
# Format the date stored in the 'date_0' variable in the '10-Mar-2004' format.
print(date_0.strftime('%D-%B-%Y'))
```

```
# Format the date stored in the 'date_0' variable in the 'March 10, 2004' format.
print(date_0.strftime('%B %d,%Y'))
```

**The `time()` Function**

To get time from a timestamp object, apply the `time()` function. It returns a `datetime.time` object.

```
# Get the time from the timestamp stored in the 'timestamp_0' variable.
time_0=timestamp_0.time()
time_0
```

**The `hour, minute` & `second` Attributes**

To get hour, minute and second values from a `datetime.time` object, apply the `hour, minute` and `second` attributes.

The output will be hour, minute and second as integer values.

```
#Get the hour, minute and second attributes from the time stored in the 'time_0' variable.
print(time_0.hour)
print(time_0.minute)
print(time_0.second)
```

# datetime Module:

- **datetime Module:** This is a general-purpose module in the Python standard library for working with dates and times. It provides classes like date, time, datetime, and timedelta to represent dates, times, timedeltas (durations), and perform basic operations like creating, formatting, and manipulating date-time objects. It's suitable for general date-time handling in Python, not just for pandas DataFrames.

- **pd.to_datetime:** This function is specifically part of the pandas library. It's used to convert strings, integers, timestamps, or other objects into pandas-specific DatetimeIndex objects. These objects are optimized for efficient time series data analysis within pandas DataFrames. They allow for advanced functionalities like indexing, resampling, date manipulation, and working with time zones, which are essential for pandas data analysis.

**Key points:**
- Use the datetime module for general date-time needs in Python, outside of pandas.
- Use pd.to_datetime when working with time series data within pandas DataFrames to leverage their optimized functionality and integration with other pandas tools.

### Activity 5: The `datetime.now()` function

The `datetime` module allows us to create date and time objects and manipulate them as we desire. Let's start with importing the `datetime` module.

```
# Import the 'datetime' module.
import datetime
```

Let's now print the current date and time using the `datetime.now()`. It returns the current time in Greenwich Mean Time (GMT) or Coordinated Universal Time (UTC) time zone.

The syntax for using the `datetime.now()` function is `datetime.datetime.now()` where the first occurrence of `datetime` denotes the module name and its second occurrence denotes the object type.

**Note:** The default format of the `datetime` object is `YYYY-MM-DD HH:MM:SS` where

- `YYYY` denotes year,

- `MM` denotes month,

- `DD` denotes day,

- `HH` denotes hours in 24-hours,

- `MM` denotes minutes, and

- `SS` denotes seconds

```
# Get the current time, print it and its data-type.
time_now = datetime.datetime.now()
print(time_now)
```

The fractional value in the above time is microseconds. That's the level of precision of a `datetime` object.

You can apply the `day, month, year, hour, minute` and `second` attributes to get the corresponding day, month, year, hour, minute and second values.

```
# Get the corresponding day, month, year, hour, minute and second values.
print(time_now.day)
print(time_now.month)
print(time_now.hour)
print(time_now.minute)
print(time_now.second)
```

### Activity 6: The `timedelta` Object

To get time in Indian Standard Time (IST) time zone, we have to add 5 hours and 30 minutes to the UTC time because IST is equivalent to UTC+5:30. For this, we need to use the `timedelta` object. It represents a duration, the difference between two dates or times.

Inside the `timedelta` object, you may pass any one of the following parameters (or arguments).

- `days`,

- `seconds=0`,

- `microseconds`,

- `milliseconds`,

- `minutes`,

- `hours`, and

- `weeks`

The default value of each of the above parameters is zero. Let's pass `hours=5` and `minutes=30` parameters to the `timedelta` object. Then add it to the current time.


```
#Get the current time in IST (or UTC+5:30) time zone. Print it and its data-type.
time_ist = time_now + datetime.timedelta(hours = 5,minutes = 30)
print(time_ist)
```

You can also convert the datetime in any time zone by simply adding or subtracting the hours and minutes values from the current time. As a practice, let's convert the current time in the UTC-7:00 zone. It's the time zone for the pacific region of the USA. The silicon valley is located in the same regions.

```
# Get the current time in the UTC-7 time zone.
time_cal = time_now + datetime.timedelta(hours = 1)
print(time_cal)
```

### Activity 8: The `date()` & `time()` Functions

Let's first separate the date and time values from the current `datetime` object using the `date()` and `time()` functions respectively.

Also, let's convert the date and time values into string and format them in the `Month Day, Year` and `HH:MM:SS AM/PM` date and time formats respectively.

```
# Separate the date and time values from the 'datetime' object. Convert them into string
b = datetime.datetime.now() + datetime.timedelta(minutes=30, hours=5)
b_date = b.date().strftime('%B %d, %Y')
b_time = b.time().strftime('%I:%M:%S %p')
print(b_date,b_time)
```

### Activity 9: The `datetime.strptime()` Function

We can create a `datetime` object by passing the string values to the `strptime()` function.

As an example, let's create a `datetime` object for the date December 26, 2024.


The `strptime()` function requires two parameters:  

- The `datetime` object as a string

- The combination of format codes for the passed `datetime` object. E.g., if the datetime object is `May 04, 2020 06:14:48 PM`, then the required combination format codes is `'%B %d, %Y %I:%M:%S %p'`

```
# Create a datetime object by passing the string values to the 'strptime()' function.
example1 = datetime.datetime.strptime('Dec 26, 2024 03:34:00', '%b %d, %Y %H:%M:%S')
print(example1)
```
