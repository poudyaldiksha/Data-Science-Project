```python
# Last class activity


# Import the required modules and load the time-series dataset on air quality
import pandas as pd

air_df=pd.read_csv("/content/drive/MyDrive/datasets/AirQualityUCI.csv", sep=";")
air_df.head()

# Removing the 'Unnamed: 15' & 'Unnamed: 16' columns.
air_df = air_df.drop(columns=['Unnamed: 15', 'Unnamed: 16'], axis=1)

# Removing the null values.
air_df = air_df.dropna()

# Creating a Pandas series containing 'datetime' objects.
time = []
for i in air_df["Time"]:
    time.append(i.replace(".",":"))

time_s = pd.Series(time)
time_s

date_time = air_df["Date"] + " " + time_s

#Convert to datetime values
date_time = pd.to_datetime(date_time, format="%d/%m/%Y %H:%M:%S")


# Remove the Date & Time columns from the DataFrame and add the 'date_time' series in it.
air_df = air_df.drop(columns=['Date', 'Time'], axis=1)
air_df.insert(loc=0, column='DateTime', value=date_time)



# Get the Pandas series containing the year values as integers.
year_series = date_time.dt.year

# Get the Pandas series containing the month values as integers.
month_series = date_time.dt.month

# Get the Pandas series containing the day values as integers.
day_series = date_time.dt.day

# Get the Pandas series containing the days of a week, i.e., Monday, Tuesday, Wednesday etc.
day_name_series = date_time.dt.day_name()

# Add the 'Year', 'Month', 'Day' and 'Day Name' columns to the DataFrame.
air_df['Year'] = year_series
air_df['Month'] = month_series
air_df['Day'] = day_series
air_df['Day Name'] = day_name_series

# Sort the DataFrame by the 'DateTime' values in the ascending order.
air_df = air_df.sort_values(by='DateTime')

air_df.head()
```

### Activity 1: commas to dot replacement

The  values in the `CO(GT), C6H6(GT), T,	RH` and	`AH` columns contain the commas in them. Let's replace the commas with periods (or dots).

```python
#replace the commas with periods in a Pandas series.
def replace_comma(b):
    list1 = []
    for i in b:
        list1.append(float(str(i).replace(",",".")))
    return pd.Series(list1)
```

```python
#Apply the 'replace_comma()' function on the ''CO(GT)', 'C6H6(GT)', 'T', 'RH' and 'AH' columns.
list2 = ['CO(GT)', 'C6H6(GT)', 'T', 'RH', 'AH' ]
for i in list2:

    air_df[i] = replace_comma(air_df[i])
```

### Activity 5: Pandas `replace` vs. String `replace` function

**Pandas replace**
1. Context: Used with pandas DataFrames or Series.
2.  Purpose: Allows you to replace specific values in your DataFrame or Series. It can handle different data types, such as integers, floats, and strings.
3.  Usage:
     -  You can replace values across the entire DataFrame or Series.
     - Supports replacing multiple values at once using a dictionary.

```python
# Create a DataFrame
df = pd.DataFrame({
    'A': [1, 2, -200, 4],
    'B': [5, -200, 7, 8]
})
df
```

```python
# Replace -200 with 8.6 in the entire DataFrame
df = df.replace(-200, 8.6)
df
```

- The replace function is called directly on the DataFrame object.
- There is no need to prefix it with `pd. `because you are calling it as a method on a DataFrame which is an object of the pandas library.
- The same principle applies to other DataFrame methods, such as `drop`, `rename`, etc. These methods are designed to be used directly on DataFrame or Series objects once pandas is imported.

**String replace**
1. Context: Used with string objects in Python.
2. Purpose: Replaces specified substrings in a string with another substring.
3. Usage:
    - Applies only to string objects and is used to modify text content.
    - Does not handle data structures like DataFrames or Series.

```python
a = ["1.2" ,"2.3"]
b =[]
for i in a:
    b.append(str(i).replace(".",","))
print(b)
```

```python
"hello.....".replace(".",",")
```

```python
# Replace 'world' with 'Python' in a string
text = "Hello, world!"
new_text = text.replace('world', 'Python')
print(new_text)
```

**Key Differences**
1. Data Type:

  - Pandas replace: Operates on pandas DataFrames or Series, which are data structures containing potentially mixed data types.
  - String replace: Operates on string objects, which are sequences of characters.
2. Functionality:

 -  Pandas replace: Can replace values across an entire DataFrame or Series, and can handle various data types (e.g., integers, floats, strings).
 - String replace: Replaces parts of the text within a single string, only applicable to strings.

 **Summary**
- Use pandas replace when you need to modify values within pandas DataFrames or Series.
- Use string replace when you need to modify specific parts of a text within a string.

### Activity 6: Garbage Value Replacement

**Why Replace Garbage Values with the Median?**

When working with data, it's common to encounter "garbage values," which can be missing, invalid, or extreme outliers that don't represent real-world phenomena. Cleaning such data is crucial for accurate analysis and modeling. One common strategy is to replace these garbage values with the median. Here’s why the median is often preferred:

**When to Use the Median for Imputation**
- When Data is Skewed: If the data distribution is skewed, the median provides a better central value than the mean.
- When Outliers are Present: In the presence of outliers, the median is more robust and less influenced by extreme values.
- For Small Datasets: In smaller datasets, the median can be a reliable measure to replace garbage values without introducing significant bias.


```python

# Example data with an outlier
data = {'values': [1, 2, 3, 4, 100]}
df = pd.DataFrame(data)

# Calculate the mean and median
mean_value = df['values'].mean()
median_value = df['values'].median()

print(f"Mean: {mean_value}, Median: {median_value}")
```

```python
b = pd.Series([1,2,3,4,5,100,100,100,100,100])
b.describe()
```

In this case, the mean is heavily influenced by the outlier `(100)`, while the median `(3)` represents the central tendency of the data more accurately.

### Activity 7: The `groupby()`  Function


The groupby function in pandas allows you to group rows in a DataFrame based on the values in one or more columns. Once grouped, you can perform various operations on each group independently, such as calculating averages, sums, or counts.


To group the values of a DataFrame by a particular value, use the `groupby()` function.

**Syntax:** `dataframe.groupby(by='column')`

You can additionally set the `sort` parameter either equal to `True` or `False` depending on whether you want to sort the values or not. By default, the `groupby()` function returns a new Pandas `DataFrameGroupBy` object containing the grouped sorted values.

```python
# Create a DataFrame
data = {
    'Region': ['North', 'South', 'North', 'South', 'East', 'West'],
    'Sales': [200, 150, 180, 120, 250, 300]
}

df = pd.DataFrame(data)
df
```

```python
# Group by 'Region' and calculate the sum of sales in each region
grouped_df = df.groupby('Region')
grouped_df
```

The `group_by()` function returns a `DataFrameGroupBy` object with its physical memory location.


**The `get_group()` Function**

To get all the occurrences of a value from a group, use the `get_group()` function.

**Syntax:** `DataFrameGroupBy_object.get_group(item)`

```python
# Get all the records for North.
grouped_df.get_group("North")
```

```python
df.groupby("Region").get_group("North").sum()
```

```python
# Group by 'Region' and calculate the sum of sales in each region
groupsales_df = df.groupby('Region').sum()
groupsales_df
```

### Activity 8: The `agg()` Function

 The aggregation functions are `count(), sum(), min(), max(), mean()` etc.


Instead of applying the aggregation functions on a `DataFrameGroupBy` object individually, you can apply them together using the `agg()` function. Inside the `agg()` function, you have to pass the names of the functions (as string values) as inputs.


The agg function in pandas is used to perform multiple aggregation operations on a DataFrame or Series after grouping data with the groupby function. It provides a flexible way to apply various aggregation functions to different columns, which can be useful for summarizing and analyzing data.



**Syntax:** `df.groupby('column_name').agg({'column1': 'agg_function1', 'column2': 'agg_function2'})`

```python
# Create a DataFrame
data = {
    'Region': ['North', 'South', 'North', 'South', 'East', 'West'],
    'Sales': [200, 150, 180, 120, 250, 300],
    'Profit': [50, 30, 40, 20, 70, 90]
}

df = pd.DataFrame(data)
df
```

```python
# Create a DataFrame
data = {
    'Region': ['North', 'South', 'North', 'South', 'East', 'West'],
    'Sales': [200, 150, 180, 120, 250, 300],
    'Profit': [50, 30, 40, 20, 70, 90]
}

df = pd.DataFrame(data)

# Group by 'Region' and calculate the sum of sales and the average profit
grouped_df = df.groupby('Region').agg({'Sales': 'sum', 'Profit': 'mean'})

print(grouped_df)
```

### Activity 11: Multivariate Grouping & Aggregation

We can also group multiple columns at once by passing a list of columns to be grouped.


 **Grouping by Multiple Columns**

 You can also group by more than one column. For example, suppose we have data on sales and profits by both region and product type:

```python
# Create a DataFrame
data = {
    'Region': ['North', 'South', 'North', 'South', 'East', 'West', 'East', 'West'],
    'Product': ['A', 'B', 'B', 'A', 'A', 'B', 'B', 'A'],
    'Sales': [200, 150, 180, 120, 250, 300, 160, 190],
    'Profit': [50, 30, 40, 20, 70, 90, 60, 80]
}

df = pd.DataFrame(data)

# Group by 'Region' and 'Product' and calculate the sum of sales and average profit
grouped_df = df.groupby(['Region', 'Product']).agg({'Sales': 'sum', 'Profit': 'mean'})

print(grouped_df)
```
